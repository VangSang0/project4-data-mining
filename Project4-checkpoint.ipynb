#### Handling imports

```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
phones_df = pd.read_csv('Mobile.csv')
```

```python
phones_df.head()
```

```python
##I had chatGPT help me write this function just because of the nature of "Product Description".

pattern = r'(\d+(\.\d+)?)MP\s+(Rear|Front)\s+Camera'

# Function to extract featured camera information
def extract_featured_camera(text):
    matches = re.findall(pattern, text)
    featured_cameras = []
    if matches:
        for match in matches:
            mp, _, camera_side = match
            featured_cameras.append(f"{mp}MP {camera_side}")
    return ", ".join(featured_cameras) if featured_cameras else None

# Apply the function to create the new column
phones_df['featured_camera'] = phones_df['Product Descriptation'].apply(extract_featured_camera)

# Display the DataFrame
print(phones_df['featured_camera'])
```

```python
phones_df['battery(mAh)'] = phones_df['battery(mAh)'].replace('not_listed', np.nan)
```

```python
not_listed_count = (phones_df['battery(mAh)'] == 'not_listed').sum()

print("Number of entries with 'not_listed' in the 'battery(mAh)' column:", not_listed_count)
```

### Further enhancing the featured_camera to add one more dimensionality 

```python

label_encoder = LabelEncoder()

phones_df['featured_camera_encoded'] = label_encoder.fit_transform(phones_df['featured_camera'])

phones_df.drop('featured_camera', axis=1, inplace=True)

print(phones_df)
```

### Starting the k-means clustering process

```python
phones_df.head()
```

```python
## I used the elbow method program from "Elbow Method for Finding the Optimal Number of Clusters in K-Means" by Basil Saji with scaling

# Select numerical features for clustering
features = ['Product_Prices', 'RAM(gb)', 'ROM(gb)', 'Screen_Size', 'battery(mAh)', 'featured_camera_encoded']
X = phones_df[features]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow Method to determine the optimal number of clusters (K)
inertia_values = []
k_values = range(1, 11)  # Try K from 1 to 10 clusters
for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow method graph
plt.plot(k_values, inertia_values, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCC')
plt.title('Elbow Method for Optimal K')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
# Implementing the KMeans when k = 3
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
kmeans.fit(X_scaled)
cluster_labels = kmeans.labels_
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, cmap='viridis', s=50)
plt.title('K-means Clustering Results (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
phones_df['Cluster'] = cluster_labels

# Plot parallel coordinates plot
plt.figure(figsize=(10, 6))
pd.plotting.parallel_coordinates(phones_df, 'Cluster', colormap='viridis')
plt.title('Parallel Coordinates Plot of Clusters')
plt.xlabel('Features')
plt.ylabel('Feature Values')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Select numerical features for clustering (excluding 'Product_Prices')
features = ['RAM(gb)', 'ROM(gb)', 'Screen_Size', 'battery(mAh)', 'featured_camera_encoded']
X = phones_df[features]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform KMeans clustering with k=3
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
kmeans.fit(X_scaled)
cluster_labels = kmeans.labels_

# Apply PCA to reduce dimensions to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Add cluster labels to DataFrame
phones_df['Cluster'] = cluster_labels

# Plot scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=cluster_labels, palette='viridis', s=50)
plt.title('Scatter Plot of Clusters (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# Plot parallel coordinates plot
plt.figure(figsize=(10, 6))
pd.plotting.parallel_coordinates(phones_df, 'Cluster', cols=features, colormap='viridis')
plt.title('Parallel Coordinates Plot of Clusters (excluding Product_Price)')
plt.xlabel('Features')
plt.ylabel('Feature Values')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
